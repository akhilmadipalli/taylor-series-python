```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import math

def newsin(angle):

    angle %= 2*math.pi
    
    result = 0
    
    sign = 1
    
    power = 1
    
    for i in range(15):
    
        result += (angle**power / math.factorial(power)) * sign
        
        sign *= -1
        
        power += 2
        
    return result
```

```python
def newcos(angle):

    angle %= 2*math.pi
    
    result = 0
    
    sign = 1
    
    power = 0
    
    for i in range(15):
    
        result += (angle**power / math.factorial(power)) * sign
        
        sign *= -1
        
        power += 2
        
    return result
```

```python
def newtan(angle):
    return newsin(angle)/newcos(angle)
```

```python
def newexp(value):

    result = 0

    power = 0
    
    for i in range(100):
    
        result += (value**power) / math.factorial(power)
        
        power += 1
        
    return result
```

```python
import numpy as np
```

```python
sinplot = np.linspace(0, 2*math.pi, 1000)
cosplot = np.linspace(0, 2*math.pi, 1000)
tanplot = np.linspace(0, math.pi, 100)
expplot = np.linspace(-3, 3, 1000)
```

```python
siny = []
for i in sinplot:
    siny.append(newsin(i))
cosy = []
for i in cosplot:
    cosy.append(newcos(i))
tany = []
for i in tanplot:
    tany.append(newtan(i))
expy = []
for i in expplot:
    expy.append(newexp(i))
```

```python
plt.plot(sinplot, siny, 'b')
plt.title('newsin')
plt.show()
```

```python
plt.plot(cosplot, cosy, 'r')
plt.title('newcos')
plt.show()
```

```python
plt.plot(tanplot, tany, 'y')
plt.title('newtan')
plt.show()
```

```python
plt.plot(expplot, expy, 'black')
plt.title('newexp')
plt.show()
```
